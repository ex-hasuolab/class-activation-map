# ImageNetからダウンロード
ImageNetDownloaderクラスの download(wnid) 関数にWNIDを渡してダウンロード  

```python
import os
from download_imagenet import ImageNetDownloader
```

```python
# ダウンローダー生成
downloader = ImageNetDownloader(root='/data/dog_imagenet')
```

## ダウンロードする画像を探す

```python
# 上位概念のWNID
wnid_dog = 'n02084071'  # 犬画像
```

```python
# すべての下位概念を含めたwnidを取得
children_wnid = downloader.wnid_children(wnid_dog, recursive=True)

# 各wnidのword（クラスラベルになるもの）と画像数を取得
children_word = {
    downloader.wnid_to_words(c)[0]:
        [
            c,
            len(downloader._get_data_with_url(downloader.IMG_LIST_URL.format(c)).split())
        ]
    for c in children_wnid}

children_word
```

```python
# 読み込みたいクラスを選択
choose_word = {
    'golden retriever': 'n02099601',
    'Labrador retriever': 'n02099712',
    'Yorkshire terrier': 'n02094433',
    'schnauzer': 'n02096756',
    'beagle': 'n02088364',
    'dalmatian': 'n02110341',
    'corgi': 'n02112826',
    'toy poodle': 'n02113624',
    'standard poodle': 'n02113799',
    'pug': 'n02110958',
    'collie': 'n02106030',
    'German shepherd': 'n02106662',
    'Doberman': 'n02107142',
    'Siberian husky': 'n02110185',
    'bulldog': 'n02108672',
    'Pomeranian': 'n02112018',
    'Chihuahua': 'n02085620'
}
```

## ダウンロード

```python
# 各クラス毎に画像を読み込む
# kerasのdata_generatorに合うようなフォルダ構成にする
for word, wnid in choose_word.items():
    img_dir = os.path.join('/data/dog_imagenet/train', word)
    downloader.download(wnid, img_dir=img_dir, list_dir='/data/dog_imagenet/list', limit=600, verbose=True, timeout=7)
```

# あるディレクトリからいくつかデータを移動する
（雑なコードなので適宜書き換えること。下記はtrainディレクトリからtestディレクトリへ少しデータを移動）

```python
import os
import glob
import shutil
```

```python
train_dir = '/data/dog_imagenet/train'
test_dir = '/data/dog_imagenet/test'
for word in choose_word.keys():
    # 各クラスのディレクトリ
    class_dir_train = os.path.join(train_dir, word)
    class_dir_test = os.path.join(test_dir, word)
    if not os.path.exists(class_dir_test):
        os.mkdir(class_dir_test)
    img_dir_list = glob.glob(os.path.join(os.path.join(train_dir, word), '*'))
    for i in range(80):  # 80個分移動
        # 画像のファイル名のみ取り出す
        img_file = img_dir_list[i].split('/')[-1]
        
        shutil.move(os.path.join(class_dir_train, img_file), class_dir_test+'/')  # 移動
```
